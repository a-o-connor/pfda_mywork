```python
phone_numbers = """123-456-7890
                    987.654.321 # an ip address
                    234-567-8901
                    654.321.987 # an ip address
                    345-678-9012
                    +321 654 9784 # a phone number with a .
                    456-789-012 # badly formatted
                    999.666.333
                    45678   # I don't know what this is !!
                """

#Find only the phone numbers: 
re.findall(r"\d{3}-\d{3}-\d{4}", phone_numbers)

#include ones with missing digit at end 
re.findall(r"\d{3}-\d{3}-\d{3,}", phone_numbers)

#include the ones with either a hyphen or a space
re.findall(r"\d{3}[\- ]\d{3}[\- ]\d{3,}", phone_numbers)
```

## Using Regex in PANDAs 
Clean up phone numbers in the people csv.

```python
import pandas as pd 

people_df = pd.read_csv("people-1000.csv")
people_df.head(10)
```

```python
#gewt rid of any extension in the phone numbers column:

the_ext_pattern = r"x\d" #an "x" followed by any number of digits 
people_df["clean_phone"] = people_df["Phone"].str.replace(the_ext_pattern, "", regex=True)
people_df.head(10)
```

```python
#get rid of any hyphens, dots etc in the phone no column 
#to match any literal characters have to preface with a backslash, because . can mean "match any character " in regex
# space does not have to be prefaced with a \
other_chars = r"[\-\. \(\)]*" #find dash,dot, space, open barck, close bracket. Match 0 or more times 
people_df["clean_phone"] = people_df["clean_phone"].str.replace(other_chars, "", regex=True)
```
